2. **Multiplication quiz**

Suppose you want to develop a program to help a first-grader practice multiplication. The program randomly generates two single-digit integers, number1 and number2, and displays to the student a question such as; What is 1 * 7. To generate a random number, you can use the randint(a, b) function in the random module. This function returns a random integer i between a and b, inclusively. To obtain a random integer between 0 and 9, use randint(0, 9).

The program may be set up to work as follows:

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
marks = [40,56,32,76,55,67,69,45]

passed=0
failed=0

for m in marks:
    if m >= 50:
        passed += 1 # tallies up how many passed
    else:
        failed += 1 # tallies how many failed

highest_mark = max(marks)
lowest_mark = min(marks)

average = sum(marks) / len(marks) # lenght of marks
print("average:",average)

print(f"Average mark: {average:}")
print(f"Highest mark: {highest_mark}")
print(f"Lowest mark: {lowest_mark}")
print(f"Number of students who passed: {passed}") # brings back that tally of passed
print(f"Number of students who failed: {failed}") # brings back tally of failed
````
